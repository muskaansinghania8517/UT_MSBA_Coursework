# Question 1: Expected Loss Minimization [10 pts]
Consider a binary classification problem with the following loss matrix -
$$
   {\begin{array}{ccccc}
   & & \text{Predicted class} & \text{           } &\\
   & & C1 & C2 & Reject\\
   \text{True class} & C1 & 0 & r & c  \\
   & C2 & s & 0 & c \\
  \end{array} } 
$$

where the cost of rejection is a constant, and the costs $r$ and $s$ are positive real numbers. Let $f(x)=P(C1|x)$.

**(b) [2.5 points]** For $c=0$, show that the decision which minimizes the expected loss is to reject all instances of $x$

## [5 points] Data Preprocessing 

(a) [1 pt] Some of the columns have values like - no internet service or no phone service. Replace these two values with with a simple No

```python
# Replace values with No
for col in ['MultipleLines','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingTV','StreamingMovies']:
    df.loc[df[col].str.contains('No', case=False), col] = 'No'
```

(b) [1 pt] Convert all 'Yes' values to 1 and 'No' values to 0 <br>
(c) [1 pt] Convert all  'Female' values to 1 and 'Male' values to 0

```python
# Changing values to 0/1

yes_no_columns = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup',
                  'DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
for col in yes_no_columns:
    df[col]=df[col].apply(lambda x: 1 if x =='Yes' else 0)
```

```python
df['gender']=df['gender'].apply(lambda x: 1 if x =='Female' else 0)
```

(d) [2 pts] Perform one-hot encoding on the folowing columns: 'InternetService', 'Contract', 'PaymentMethod'

```python
# One hot encoding for categor ical columns
# Name your final dataframe df 2

# Your code here

df2 = pd.get_dummies(df, columns=['InternetService', 'Contract', 'PaymentMethod'])
```

```python
# Scaling
cols_to_scale = ['tenure','MonthlyCharges','TotalCharges']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale])
```

```python
df2.head()
```

e) [1 pt] Split the dataset into training and test using a 80-20 split with a random state of 15. Perform stratified sampling on the target variable while doing so. Split the training data further into training (80%) and validation sets (20%) with a random state of 15. Perform stratified sampling on the target variable while doing so.

Note: The 'stratify' parameter can be used for this in the train_test_split function. 

```python
# Train test split
X = df2.drop(['Churn'],axis=1)
y = df2['Churn']
#stratification means that the train_test_split method returns training and test subsets that have the same proportions of class labels as the input dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=15, stratify=y)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.20, random_state=15, stratify=y_train)
```

```python
# With samples corresponding to the positive class being very low, we can clearly see the imbalance in our data 
print('Churn occurences in the training set \n', y_train.value_counts())
print('\n')
print('Churn occurences throughout the data \n', y.value_counts())
```

```python
# Decision Tree Classifier
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix ,ConfusionMatrixDisplay
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(20,5))
clf = DecisionTreeClassifier(random_state=0,criterion="entropy", max_depth=3)
clf = clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
y_pred_val=clf.predict(X_val)
ConfusionMatrixDisplay.from_predictions(y_val,y_pred_val, display_labels=clf.classes_,ax=ax1)
ax1.set_title('Validation Data')
ConfusionMatrixDisplay.from_predictions(y_test,y_pred, display_labels=clf.classes_,ax=ax2)
ax2.set_title('Test Data')
plt.show()
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(20,5))
clf_lr = LogisticRegression()
clf_lr = clf_lr.fit(X_train, y_train)
y_pred_lr = clf_lr.predict(X_test)
y_pred_val_lr=clf_lr.predict(X_val)
ConfusionMatrixDisplay.from_predictions(y_val,y_pred_val_lr, display_labels=clf.classes_,ax=ax1)
ax1.set_title('Validation Data')
ConfusionMatrixDisplay.from_predictions(y_test,y_pred_lr, display_labels=clf.classes_,ax=ax2)
ax2.set_title('Test Data')
plt.show()
```

```python
#Finding the hyperparameters
from sklearn.neural_network import MLPClassifier
mlp_gs = MLPClassifier(max_iter=100)
parameter_space = {
    'hidden_layer_sizes': [(10,30,10),(20,)],
    'activation': ['tanh', 'relu'],
    'solver': ['sgd', 'adam'],
    'alpha': [0.0001, 0.05],
    'learning_rate': ['constant','adaptive'],
}
from sklearn.model_selection import GridSearchCV
clf_GV = GridSearchCV(mlp_gs, parameter_space, n_jobs=-1, cv=5)
clf_GV.fit(X_train, y_train)
```

```python
print('Best parameters found:\n', clf_GV.best_params_)
```

```python
from sklearn.neural_network import MLPClassifier
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(20,5))
clf_MLP = MLPClassifier(activation='tanh', alpha=0.05, hidden_layer_sizes= (10, 30, 10),learning_rate='constant', solver= 'adam').fit(X_train, y_train)

y_pred_MLP = clf_MLP.predict(X_test)
y_pred_val_MLP=clf_MLP.predict(X_val)
ConfusionMatrixDisplay.from_predictions(y_val,y_pred_val_MLP, display_labels=clf.classes_,ax=ax1)
ax1.set_title('Validation Data')
ConfusionMatrixDisplay.from_predictions(y_test,y_pred_MLP, display_labels=clf.classes_,ax=ax2)
ax2.set_title('Test Data')
plt.show()
```

```python

# Plot ROC curves
from sklearn.metrics import roc_curve, roc_auc_score

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(20,5))

metrics.RocCurveDisplay.from_estimator(clf, X_val, y_val,ax=ax1)
metrics.RocCurveDisplay.from_estimator(clf_lr, X_val, y_val,ax=ax1)
metrics.RocCurveDisplay.from_estimator(clf_MLP, X_val, y_val,ax=ax1)
#add axis labels to plot
ax1.set_title('Validation Data')

metrics.RocCurveDisplay.from_estimator(clf, X_test, y_test,ax=ax2)
metrics.RocCurveDisplay.from_estimator(clf_lr, X_test, y_test,ax=ax2)
metrics.RocCurveDisplay.from_estimator(clf_MLP, X_test, y_test,ax=ax2)
ax2.set_title('Test data')
plt.show()
```

```python
# Plot PR curves
from sklearn.metrics import precision_recall_curve, PrecisionRecallDisplay

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(20,5))

metrics.PrecisionRecallDisplay.from_estimator(clf, X_val, y_val,ax=ax1)
metrics.PrecisionRecallDisplay.from_estimator(clf_lr, X_val, y_val,ax=ax1)
metrics.PrecisionRecallDisplay.from_estimator(clf_MLP, X_val, y_val,ax=ax1)
ax1.set_title('Validation Data')

metrics.PrecisionRecallDisplay.from_estimator(clf, X_test, y_test,ax=ax2)
metrics.PrecisionRecallDisplay.from_estimator(clf_lr, X_test, y_test,ax=ax2)
metrics.PrecisionRecallDisplay.from_estimator(clf_MLP, X_test, y_test,ax=ax2)
ax2.set_title('Test data')
plt.show()
```

```python
from sklearn.metrics import f1_score
print("Logit: Validation : F1 Score: " ,f1_score(y_val, y_pred_val_lr))
print("Logit: Test : F1 Score: " ,f1_score(y_test, y_pred_lr))
print("\n MLP: Validation : F1 Score: " ,f1_score(y_val, y_pred_val_MLP))
print("MLP: Test : F1 Score: " ,f1_score(y_test, y_pred_MLP))
```
