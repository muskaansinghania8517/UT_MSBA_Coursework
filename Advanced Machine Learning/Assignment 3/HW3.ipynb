# Answer:
t-SNE is a dimensionality reduction method specially used when the data in higher dimensions cannot be divided using linear properties. Also, unlike PCA, t-SNE cannot be used for Machine Learning as the results produced from t-SNE cannot be reproduced. Infact, since t-SNE is iterative the same data might get different results when run multiple times. 

t-SNE works on the principle of calculating the probability of a point selecting another point as its neighbour. These probabilities should be proportional to the probabilities calculated from the gaussian distribution made from the distances calculated from the point in question to other point. But, in t-SNE Student distribution is used because gaussian distribution has a short tail and we need to have as less number of points as possible with high probability of being a neighbour to the point in question i.e., the distribution needs to have a long tail.  


# Question 3: PCA (8 pts)

Explain the principle of Principal Component Analysis algorithm and in what sense is the selection of the eigenvectors to represent the data is an optimal choice.  How do you reconstruct (a noisy version of) the original data from the eigenvectors and the loadings?



# Answer:
Principal Component Analysis works on the principle of constructing new variables using linear combinations with the initial features. For example, for an n-dimensional feature space we get n principal components. The point of principal components is that we choose those components which carry the maximum variation in the data. To identify the principal components which cover most of the information of the data we calculate the eigen values of the eigen vectors for the covariant matrix ($A$) constructed from the feature space. First, the eigen vectors $\lambda$ are calculated from the below determinant equation.

$$|A - \lambda I| = 0 $$

$$I = Identity \space Matrix$$

$$A = Covariant \space Matrix = \begin{bmatrix}
\sigma(x_1,x_1) & \sigma(x_1,x_2) & \sigma(x_1,x_3)\\
\sigma(x_2,x_1) & \sigma(x_2,x_2) & \sigma(x_2,x_3)\\
\sigma(x_3,x_1) & \sigma(x_3,x_2) & \sigma(x_3,x_3)
\end{bmatrix}$$  

$$\sigma(x_i,x_i) = variance \space in \space x_i \space and \space \sigma(x_i,x_j) \space = \space covariance \space between \space x_i \space and \space x_j\$$

The eigen values are then used to get the corresponding eigenvectors by using the following equation:

$$ (A - \lambda I)X = 0 $$

where $X$ is the corresponding eigenvector

The eigenvectors whose eigen values are high are selected as those with high eigen values are the ones which cover most of the variance in the feature space. This is because the eigenvectors and eigen values are calculated from the covariance matrix of the feature space as explained above.

The Principal Component values are calculated by multiplying the matrix of eigen vectors with the feature space. To reconstruct the original data we need multiply the eigenvectors matrix inverse with Principal Component values. If the information of all Principal Component values is not available then we get a noisy version of the original data.

$$ PC = X.V$$
$$PC = Principal \space Component \space values \space matrix, \space X = scaled \space data, \space V = eigenvector \space matrix $$
$$X_{reconstructed} = PC.V^T$$

e) i)**(2pts)** Calculate 95% confidence intervals for SalePrice with a sample size of 100. 

ii)**(3pts)** Calculate 95 % confidence intervals for 100 different trials with a sample size of 500. Plot the confidence intervals and interpret how it captures the population mean.

````python
import scipy.stats

np.random.seed(10)
sample_size = 100
sample = np.random.choice(a= data_dm['SalePrice'],
                          size = sample_size)
sample_mean = sample.mean()

#Get the critical Z value
z_critical = scipy.stats.norm.ppf(1-0.05/2)

#Get population standard deviation
pop_stdev = sample.std()


#margin of error
margin_of_error = z_critical * (pop_stdev/math.sqrt(sample_size)) 

#confidence interval
confidence_interval = (sample_mean - margin_of_error,
                       sample_mean + margin_of_error)  

#Print confidence interval and true mean value
print("95% Confidence Interval:")
print(confidence_interval)
print("\n")
print("True Mean Value:")
print(data_dm['SalePrice'].mean())
````

````python
np.random.seed(12)

#define the sample size
sample_size = 500
intervals = []
sample_means = []

for sample in range(100):
    sample = np.random.choice(a= data_dm['SalePrice'], size = sample_size)
    sample_mean = sample.mean()
    sample_means.append(sample_mean)
    #compute z critical value
    z_critical = scipy.stats.norm.ppf(1-0.05/2)
    #compute population std dev        
    pop_stdev = data_dm['SalePrice'].std()
    
    margin_of_error = z_critical * (pop_stdev/math.sqrt(sample_size))
    confidence_interval = (sample_mean - margin_of_error,
                           sample_mean + margin_of_error)  
    
    intervals.append(confidence_interval)
    

plt.figure(figsize=(13, 9))

plt.errorbar(x=np.arange(0.1, 100, 1), 
             y=sample_means, 
             yerr=[(top-bot)/2 for top,bot in intervals],
             fmt='o')

plt.hlines(xmin=0, xmax=25,
           y=data['SalePrice'].mean(), 
           linewidth=2.0,
           color="red")
plt.title('Confidence Intervals for 100 Trials', fontsize = 20)
plt.show()
````
