## (a). (4 pts) Describe the difference between model bias and the bias of a point estimator. 

**Model Bias**
Model Bias is the sum  of the difference between the predicted and observed values for all the observations. It describes how well a model fits the training set. A model with high bias won't match the data set closely, while a model with low bias will match the data set very closely. Models with high bias are overly simple and fail to capture the trends present in the data set whereas Models with very low bias are overly overfitted and complex models which captures each and every noise of the dataset

Bias=E[ˆθ]−θ.


**Bias of a point estimator**
A point estimator of some population parameter θ is a single numerical value of a statistic.Point estimators are defined as the functions used to extrapolate a rough estimate of a population parameter from samples of the population taken at random. To find a point estimate or statistic that serves as the best estimate of an unknown parameter of a population, they use sample data from the population.

**For example** - when we want to find the average weight of peoples attending a conference, it will be impossible to collect the exact weight of every conference peoples in the world. Instead, a statistician can use the point estimator to estimate the population parameter.

The difference between the expected value of the estimator and the value of the parameter being estimated of a population say **Mean or Standard Deviation** is referred to as **the bias of a point estimator**. Therefore, the estimator is considered unbiased when the estimated value of the parameter and the value of the parameter being estimated is equal.

Let θ^ be a point estimator of a population parameter θ.

Bias: The difference between the expected value of the estimator E[θ^] and the true value of θ, i.e.   
**Bias(θ^)=E[θ^]−θ   
When E[θ^]=θ, θ^ is called an unbiased estimator.**



## b) (6 pts). How can you use a learning curve to determine whether a model is overfitting  (for a given sample size)? Discuss this with respect to the observed train and validation error curves. How does your answer change if the model you are trying to determine if the model is underfitting?

By gradually introducing fresh training examples, learning curves plot the training and validation loss of a sample of training instances. We can determine using learning curves whether adding more training instances would increase the validation score (score on unseen data).


For an Overfit Model:

**For an overfit model for N number of observations, Training loss will be very less but the validation error will be high**

Learning curve has a **very low training loss at the beginning** which gradually increases very slightly upon adding training examples and doesn’t flatten. Learning curve has a **high validation loss at the beginning** which gradually decreases upon adding training examples and doesn’t flatten, indicating addition of more training examples can improve the model performance on unseen data.

**A plot of learning curves shows overfitting if:**

A) The plot of training loss continues to decrease with experience or increases very slightly upon adding training examples and doesnt flatten.  
B) The plot of validation loss decreases to a point and begins increasing again.

For an Underfit Model:

**For an underfit model for N number of observations, Training loss will be very high, also the validation error will be high**

Learning curve has a **high training loss at the beginning** which gradually increases upon adding training examples and suddenly falls to an arbitrary minimum point (minimum doesn’t mean zero loss) at the end. Learning curve has a **high validation loss at the beginning** which gradually lowers upon adding training examples and suddenly falls to an arbitrary minimum at the end (this sudden fall at the end may not always happen, but it may stay flat), indicating addition of more training examples can’t improve the model performance on unseen data

**A plot of learning curves shows underfitting if:**

The training loss remains flat regardless of training.  
The test loss continues to decrease until the end of training.

```python
def plot_curves(x_train, y_train, x_test, y_test):

    degree=list([2,4,7])
    x_all_pred_df = pd.DataFrame(columns=['X','Deg_2','Deg_4','Deg_7'])
    test_df=pd.DataFrame(columns=['X','Deg_2','Deg_4','Deg_7'])
    test_df['X']=pd.Series(x_test)
    x_all_pred_df['X']=pd.Series(x_all.ravel())
    test_mse_list=[]
    train_mse_list=[]
    for i in degree:    
        colname = 'Deg_'+str(i)
        print("Running for polynomial of degree",(i))
        train_fit=lrp.fit(np.vander(np.squeeze(x_train), (i+1)),y_train)

        train_pred=lrp.predict(np.vander(np.squeeze(x_train), (i+1)))
        test_pred=lrp.predict(np.vander(np.squeeze(x_test), (i+1)))
        x_all_pred=lrp.predict(np.vander(np.squeeze(x_all), (i+1)))

        train_mse=mean_squared_error(y_train,train_pred)
        test_mse=mean_squared_error(y_test,test_pred)
        
        train_mse_list.append(train_mse)
        test_mse_list.append(test_mse)

        print("MSE for train data is",train_mse)
        print("MSE for test data is",test_mse)

        test_df[colname]=test_pred
        x_all_pred_df[colname]=x_all_pred
        print("--------------------------------------------------------")
    test_df.sort_values(by="X", inplace=True)
    x_all_pred_df.sort_values(by="X", inplace=True)
    return train_mse_list,test_mse_list,test_df,x_all_pred_df
```

## 1. (5 pts) Read this [blog](https://medium.com/optimization-algorithms-for-deep-neural-networks/gradient-descent-with-momentum-dce805cd8de8) on medium and describe in your own words how momentum leads to a faster convergence of the loss function. 

**Problems of Gradient Descent**:   
a) There are chances of big vertical oscilations which can throw the GD out of the way and cost oscillates around the minima if we use high learning rate and  if the learning rate is too small, the weights will be updated very slowly hence convergence takes a lot of time even when the gradient is high

b) Incase of multiple local minima and one global minima, there can be high chances of getting stuck at local minima since the gradient gets 0 at a local minima too(at saddle points( plateau) GD is negligible or zero, which in turn leads to small or no weight updates. Hence, the network becomes stagnant, and learning stops, and even after multiple tries, it wont be able to escape the same and therefore would never be able to reach the global minima. In cases like this we get stuck with a sub optimal solution assuming it to be the optimal one.

**Gradient descent with Momentum**   
Gradient descent with momentum helps us in escaping the saddle points through momentum by taking all the previous gradients into considerations while calculating the weight update parameter. We can deduce that higher β will accommodate more gradients from the past.

Also It enables our algorithm to take a straighter forward path to local optima and to damp out vertical oscillations. The way it does it so:

In GD with Momentum, we update pur parameters as below:

VdW(t) = β * VdW(t-1) + (1 — β) * dW(t)   
AFter putting the values in Vdw(t-1), we get the following generalized equation for Vdw(t):   
VdW(t)= (1-β) * (Sum from (t=1 to t=n)((β ** n-t )* dW(t))  

**W(t) = W — learning rate * VdW(t)**    

**Case 1: When all the past gradients have the same sign**   
Since all the past gradients are of same sign, VdW(t) will be a high value,thus increasing the momentum and accelerating the descent.In this case even if we use low learning rate, our GD will be taking bigger steps, hence converging faster toward the minima. 

**Case 1: When the past gradients have opposite signs**   
Since the past gradients have different signs, VdW(t) will be a low value, thus decreasing the momentum and decreasing the acceleration of the descent making it close to 0. If the learning rate is high, the gradient at each iteration around the valley C will alter its sign between +ve and -ve and after few oscillations, the sum of past gradients will become small. Thus, making small updates in the weights from there on and hence our GD will be taking almost negligible steps, hence damping out vertical oscillations.

**Conclusion**  
Gradient Descent with Momentum helps in taking bigger steps wven with lower learning rate when the past gradients are in the same direction, whereas it damps the vertical oscillations after few steps even if with higher learning rate when the gradient keeps on oscillating around the minima.

**IN OTHER WORDS:**
Gradient Descent with Momentum takes small steps in directions where the gradients oscillate and take large steps along the direction where the past gradients have the same direction(same sign).

Gradient Descent is used to find the optimal parameter values for which the loss function is Minimum. In General Gradient Descent, every epoch is calculated on the entire dataset which can be time consuming as well as wont be computationally heavy.

One variation of Gradient Descent is Mini batch Gradient Descent, in which we take n samples out of data set of p size where n<<(P) and computes the value in every epoch only on those n samples. These n samples are randomly chosen in every epoch.

**The advantages of mini batch gradient descent are as below**:

2.(**20 pts**) Stochastic gradient descent coding <br>

Code an SGD solution in Python for this non-linear model$$ y = w_0 + w_1e^{-x_1} + w_2x_1 + w_3x_1x_2 $$  The template of the solution class is given. The init function of the class takes as input the learning rate, regularization constant and number of epochs. The fit method must take as input X, y. The predict method takes an X value (optionally, an array of values). 

a) (**15 pts**) Use the expression derived in part 1 to predict the data given in 'SGD_samples.csv', for 15 epochs, using learning rates: [0, .0001, .001, .01, 0.1, 1, 10, 100] and regularization (ridge regression) constants: [0,10,100]. For the best 2 combinations of learning_rate and regularization for SGD, plot MSE and the $w$ parameters as a function of epoch (for 15 epochs) . 



```python
%matplotlib inline
import random
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import Ridge
from tabulate import tabulate

class Regression:
    
    def __init__(self, learning_rate, regularization, n_epoch):
        self.learning_rate = learning_rate
        self.n_epoch = n_epoch
        self.regularization = regularization
        self.coef = np.zeros(4)
        
    def sgd(self, gradient):
        self.coef = self.coef - (self.learning_rate*gradient) 
    def fit(self, X, y, update_rule='sgd', plot=False):
        mse = []
        coefs = []
        X = self.get_features(X)
        for epoch in range(self.n_epoch):
            for i in range(X.shape[0]):
                
                X_i = X[i,:]
                error = self.linearPredict(X_i) - y[i]
                
                gradient = (error*X_i) + (self.regularization)*(self.coef)
         
                self.sgd(gradient)

            coefs.append(self.coef)
            residuals = y - self.linearPredict(X)         
            mse.append(np.mean(residuals**2))

        self.lowest_mse = mse[-1]
        if plot == True:
            plt.figure()
            plt.plot(range(self.n_epoch),mse)
            plt.xlabel('epoch')
            plt.ylabel('MSE')
            plt.figure()
            coefs = np.array(coefs)
            plt.plot(range(self.n_epoch),coefs[:,0],label='w0')
            plt.plot(range(self.n_epoch),coefs[:,1],label='w1')
            plt.plot(range(self.n_epoch),coefs[:,2],label='w2')
            plt.plot(range(self.n_epoch),coefs[:,3],label='w3')
            plt.legend()
            plt.xlabel('epoch')
            plt.ylabel('parameter value')

    def get_features(self, X):
        x = np.zeros((X.shape[0], 4))
        x[:,0] = 1
        x[:,1] = X[:,0]
        x[:,2] = X[:,0]*X[:,1]
        x[:,3] = np.exp(-X[:,0])
        
        return x
        
    def linearPredict(self, X):  
        return X.dot(self.coef) 
```

```python
data = pd.read_csv('SGD_samples.csv')
X = np.array([data['x1'].values, data['x2'].values]).T
y = data['y'].values
n_epochs = 15
learning_rate = [0.0001, 0.001, 0.01, 0.1, 1, 10, 100]
regularization = [0, 10, 100]
```

```python
mse_list = []
mse_least_scores = []
best_mse = []

for lr in range(0, len(learning_rate)):
    for reg in range(0, len(regularization)):
        mse_list.append((learning_rate[lr], regularization[reg]))

for i in range(0, len(mse_list)):
    sgd = Regression(mse_list[i][0], mse_list[i][1], n_epochs)
    sgd.fit(X,y,plot=False)
    mse_least_scores.append(sgd.lowest_mse)

final = np.argsort(np.array(mse_least_scores))[:2]

for i in final:
    sgd = Regression(mse_list[i][0], mse_list[i][1], n_epochs)
    sgd.fit(X,y,plot=True)
    best_mse.append(sgd.lowest_mse)
```

```python
print("The two best combinations of learning rate and regularization constant are\n",tabulate([mse_list[final[0]], mse_list[final[1]]], headers=['learning rate', 'regularization']))

print('Respective MSE scores : ', best_mse)
```
